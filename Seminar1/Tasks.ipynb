```python
import re
import pandas as pd

with open('alice.txt', 'rt', encoding='utf-8') as f:
    text = f.read()
    
text = re.sub(r'[^\w]', ' ', text)
words = text.split()
words_counts = {}
for word in words:
    if word in words_counts:
        words_counts[word] += 1
    else:
        words_counts[word] = 1
        
sorted_words_counts = sorted(words_counts.items(), key = lambda x : x[1], reverse=True)
df = pd.DataFrame(sorted_words_counts[:20], columns=["Word", "Count"])
df.style.set_properties(subset=["Word"], **{'text-align': 'left'})
```

```python
import pickle

with open('em16_edm_v00.tf', 'rt', encoding='utf-8') as f:
    data = f.read()

pattern = re.compile(r'\\begindata\s+FRAME.*?\\begintext', re.M | re.S)
matches = pattern.findall(data)

def take_info(string):
    pattern = re.compile(r'''NAME\s+=\s\'(?P<name>\w+)\'
    .*?CLASS\s+=\s+(?P<class>\d+)
    .*?CLASS_ID\s+=\s+(?P<class_id>[-\d]+)''', re.M | re.S)
    matches = pattern.search(string)
    
    return matches.groupdict()

frames_infos = list(map(take_info, matches))
frames_names = list(map(lambda x: x['name'], frames_infos))
print(frames_names)
print("Count: ", len(frames_names))

with open('dump.bin', 'wb') as f:
    pickle.dump(frames_infos, f)
    
with open('dump.bin', 'rb') as f:
    print(pickle.load(f))
```
